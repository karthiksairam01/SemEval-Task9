```python
import pandas as pd

# Assuming the file is in your current working directory.
# If not, provide the correct path.
df = pd.read_csv('food_recall_incidents.csv')

# Print some info to verify import
#print(df.head())
#print(df.info())

# Extract unique hazards and their counts
unique_hazards = df['hazard'].unique()
num_unique_hazards = len(unique_hazards)

print(f"Number of unique hazards: {num_unique_hazards}")
print("Unique hazards:")
for hazard in unique_hazards:
  print(hazard)
```

```python
# Title and language columns for those hazards that occur only once in the dataset

# Count occurrences of each hazard
hazard_counts = df['hazard'].value_counts()

# Filter for hazards that occur only once
single_occurrence_hazards = hazard_counts[hazard_counts == 1].index

# Get the rows corresponding to these hazards
single_occurrence_df = df[df['hazard'].isin(single_occurrence_hazards)]

# Print the 'title' and 'language' columns for these rows
single_occurrence_df[['title', 'language', 'hazard', 'product-category']]
```

```python
# Some 'de' language titles

# Filter for German language titles
german_titles = df[df['language'] == 'de']['title']

# Print the German titles
print("\nGerman Titles:")
for title in german_titles:
  print(title)
```

```python
# Title and language columns for those product-category that occur only once in the dataset

import pandas as pd

# Assuming the file is in your current working directory.
# If not, provide the correct path.
df = pd.read_csv('food_recall_incidents.csv')

# Count occurrences of each product category
product_category_counts = df['product-category'].value_counts()

# Identify product categories that occur only once
single_occurrence_categories = product_category_counts[product_category_counts == 1].index

# Filter the DataFrame for these categories
single_occurrence_df = df[df['product-category'].isin(single_occurrence_categories)]

# Print the 'title' and 'language' columns for the filtered data
print(single_occurrence_df[['title', 'language']])
```

```python
# Scouring the dataset to try finding unusual things

import pandas as pd

df = pd.read_csv('food_recall_incidents.csv')

# Check for missing values
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)

# Check for duplicate rows
duplicate_rows = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicate_rows}")

# Inconsistent data types
print("\nData types of each column:\n", df.dtypes)

#Further analysis:

# Check the distribution of values in categorical columns
for col in df.select_dtypes(include=['object']):
    print(f'\nValue counts for {col}:')
    print(df[col].value_counts(dropna=False))

```

```python
# Pie chart for Value counts for product-category. Whichever categories are less than 10%, grouped into one bin called "Other categories"

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('food_recall_incidents.csv')

# Calculate value counts for 'product-category'
product_category_counts = df['product-category'].value_counts(normalize=True) * 100

# Identify categories with less than 10%
other_categories = product_category_counts[product_category_counts < 10]

# Group the 'other' categories
df.loc[df['product-category'].isin(other_categories.index), 'product-category'] = 'Other categories'

# Recalculate the value counts for the updated dataframe
product_category_counts_updated = df['product-category'].value_counts(normalize=True) * 100


# Create the pie chart
plt.figure(figsize=(10, 8))
plt.pie(product_category_counts_updated, labels=product_category_counts_updated.index, autopct='%1.1f%%', startangle=90)
plt.title('Product Category Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Details of the titles having 'Recall Notification: " in them, prints its size too

import pandas as pd

# Assuming the file is in your current working directory.
# If not, provide the correct path.
df = pd.read_csv('food_recall_incidents.csv')

# Filter titles containing "Recall Notification:"
recall_notifications = df[df['title'].str.contains('Recall Notification:', na=False)]

# Print details and size
print(recall_notifications[['title', 'hazard', 'product-category']]) # Print relevant columns
print("\nSize:", len(recall_notifications))
```
